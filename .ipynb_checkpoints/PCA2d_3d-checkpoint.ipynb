### `Implementing PRINCIPAL COMPONENT ANALYSIS`

```python

def covariance_find(data):
    ## This function returns covariance matrix for the dataset
    average = np.mean(data, axis=0)
    new_data = data-average
    
    transposed_value = np.transpose(new_data)
    result = np.cov(transposed_value)
    return result

def Eigen_Finder(data_cov):
    ##  This function basically returns eigen values and eigen vectors
    val1, val2 = EGV(data_cov)
    val2 = np.transpose(val2)
    
    pos_select = np.argsort(val1)[::-1]
    
    va11 = val1[pos_select]
    val2 = val2[pos_select]
    return val1, val2
    

def principal_components(num_of_components, data):
    ##Actual function implementing PCA and returning principal components along with the 
    ### data  being projected along those components
    data_cov = covariance_find(data)
    e_val, e_vec = Eigen_Finder(data_cov)
    
    principal_components = e_vec[0: num_of_components]
    
    transposed_principal_comp = np.transpose(principal_components)
    
    val_mean = np.mean(data)
    subtracted = data-val_mean
    product_dot = np.dot(subtracted, transposed_principal_comp)
    
    return principal_components, product_dot
```

```python
#test_run
pca_comp, proj = principal_components(2,df1.drop('quality', axis=1))
print(len(pca_comp))
print(len(proj))
pca_comp
```

### We implemented PCA from scratch
#### We defined a function `principal_components()` taking two parameters namely, the number of components that we do want and the actual data itself.
#### This function uses 2 subroutines namely, `covariance_find()` and `Eigen_Finder()`. The former retruns covariance matrix for the data, and the later returns eigen values and eigen vectors(respectively, to be precise)

## PART B

### Now we have to visualize 2D and 3D pca plots for both the datasets for the `red wine` and the `white wine`
### There would be, thus, 2 plots  for both the datasets.

```python
#creating a generic function to plot 2d PCA
def plot_in_2d(first_comp, second_comp, target, size):
    plt.figure(figsize=size)
    plt.scatter(first_comp, second_comp, c=target, cmap='viridis')
    plt.xlabel('1st Principal Component')
    plt.ylabel('2nd Principal Component')
```

```python
# we have red wine dataset in df1, and white wine dataset in df2
## RED RED WINE
xred = df1.drop('quality', axis=1)
xred = xred.values
```

```python
xred = SS().fit_transform(xred)##standardized the data
```

```python
principal_comp , projection = principal_components(num_of_components=2, data=xred)
projection
```

```python
#values along the two components
comp1_values = projection[:,1:]
comp2_values = projection[:,:1]
```

```python
target = df1['quality']
size = (10,5)
plot_in_2d(comp1_values, comp2_values, target, size )
plt.title('PCA for RED WINE DATASET')
```

```python
## Similarly we visualize for the white wine dataset
xwhite = df2.drop('quality', axis=1)
xwhite = xwhite.values
```

```python
xwhite = SS().fit_transform(xwhite)
```

```python
principal_comp , projection = principal_components(num_of_components=2, data=xwhite)
comp1_values = projection[:,1:]
comp2_values = projection[:,:1]
print(projection)
```

```python
target = df2['quality']
size = (10,5)
plot_in_2d(comp1_values, comp2_values, target, size )
plt.title('PCA for WHITE WINE DATASET')
```

### Let's do 3-D visualization

```python
## Creating a generic function to plot 3d data along the 3 principal components
def plot_in_3d(first, second, third, target, size):

    figr = plt.figure(figsize=size)
    obj_axis = figr.add_subplot(111, projection='3d')
    obj_axis.scatter(first, second, third, c=target)
    
    obj_axis.set_xlabel('Component 1')
    obj_axis.set_ylabel('Component 2')
    obj_axis.set_zlabel('component 3')
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
## fitting red wine for PCA
## xred is already standardized
principal_comp_3d, projection_3d = principal_components(3, xred)
```

```python
plot_in_3d(projection_3d[:,0], projection_3d[:,1], projection_3d[:,2], df1['quality'], (10,10))
plt.title("RED WINE DATASET ")
```

```python
## plotting the 3d pca for our white wine  dataset
## the white wine dataset has already been standardized and loaded into 
## xwhite
principal_comp_3d, projection_3d = principal_components(3, xwhite)
print(projection_3d.shape)
plot_in_3d(projection_3d[:,0], projection_3d[:,1], projection_3d[:,2], df2['quality'], (10,10))
plt.title("WHITE WINE DATASET ")
```

### The PCA that we implemented from scratch has been used to plot 2d as well as 3d plots for both the datasets 'red wine' and 'white wine'
### Also the data were standardized and used to  calculate PCA and the plotting function (for both 2d as well as 3d ) has been made generic

#### Let's use PCA from sklearn.

```python
from sklearn.decomposition import PCA
```

```python
## Plotting 2-d plot for xred(red wine dataset)
pca_red = PCA(n_components=2)
comp_principal_red = pca_red.fit_transform(xred) ##xred has already been standardized
comp_principal_red
```

```python
target = df1['quality']
size = (10,5)
plot_in_2d(comp_principal_red[:,1:], comp_principal_red[:,:1], target, size )
```

```python
#plotting 3d plots
pca_red3d = PCA(n_components=3)
comp_principal_red3d = pca_red3d.fit_transform(xred) ##xred has already been standardized
comp_principal_red3d
```

```python
## calling our generic function plot_in_3d()

plot_in_3d(comp_principal_red3d[:,0],comp_principal_red3d[:,1], comp_principal_red3d[:,2]
           , df1['quality'],(10,10))
plt.title("RED WINE PLOT ")
```

```python
## 2d plot For white wine
pca_white = PCA(n_components=2)
comp_principal_white = pca_white.fit_transform(xwhite)
comp_principal_white
```

```python
## Let's plot the 2d plot
target = df2['quality']
size = (10,5)
#calling the generic function for plotting that we defined earlier
plot_in_2d(-1*comp_principal_white[:,1:], comp_principal_white[:,:1], target, size)
```

```python
## 3d plot for white wine dataset
## again the standardized white wine dataset has been stored in xwhite
pca_white3d = PCA(n_components=3)
comp_principal_white3d = pca_white3d.fit_transform(xwhite)
comp_principal_white3d.shape
```

```python
## Again calling our generic plot_in_3d function
plot_in_3d(comp_principal_white3d[:,0],comp_principal_white3d[:,1], comp_principal_white3d[:,2]
           , df2['quality'],(10,10))
plt.title("WHITE WINE PLOT")
```
